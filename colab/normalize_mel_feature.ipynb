```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv(root + "mel_mean_feature.csv")

df
```

```python
X = df.iloc[:, :128]

minmax_scaler = MinMaxScaler()
stdard_scaler = StandardScaler()

minmax_scaler.fit(X)
stdard_scaler.fit(X)

print(minmax_scaler.data_max_)
print(minmax_scaler.data_min_)
print(stdard_scaler.mean_)

minmaxX = minmax_scaler.transform(X)
stdardX = stdard_scaler.transform(X)
```

```python
minmaxdf = pd.DataFrame(minmaxX)
minmaxdf["label"] = df["label"]

minmaxdf.to_csv(root + "mel_mean_feature_minmax.csv", index=False)
```

```python
stdarddf = pd.DataFrame(stdardX)
stdarddf["label"] = df["label"]

stdarddf.to_csv(root + "mel_mean_feature_stdard.csv", index=False)
```
