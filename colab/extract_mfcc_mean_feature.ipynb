```python
train_df = pd.read_csv(root + "train.csv")

train_df
```

```python
def mfcc_mean_feature(path):
    data, sample_rate = librosa.load(path, res_type="kaiser_fast")

    mfcc = np.mean(librosa.feature.mfcc(y=data, sr=sample_rate, n_mfcc=13).T, axis=0)

    return np.array(mfcc)
```

```python
X = []
for path in train_df['path']:
    feature = mfcc_mean_feature(path)

    X.append(feature)
```

```python
mfcc_mean_feature_df = pd.DataFrame(X)

mfcc_mean_feature_df["label"] = train_df["label"]

mfcc_mean_feature_df.to_csv(root + "mfcc_mean_feature.csv")
```
